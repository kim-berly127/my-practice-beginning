```python
# 将句子/词语按照出现的频次进行排序，由于用到的次数较多，因此包装成函数
def num_sorted(dict):  
    lists = []
    for key in dict:  # 字典转换成列表
        list1 = []
        list1.append(key)  # 添加句子/词语
        # print("key",key)
        list1.append(dict[key])   # 添加句子/词语对应的出现次数
        # print("dict[key]",dict[key])
        lists.append(list1)
        # print("list1",list1)
    indices = sorted(lists, key=lambda x: x[1], reverse=True)  # 将列表中内容按照频次降序排列
    # print("lists",lists)
    return indices
indices = num_sorted(sentence_dict)
# print(indices)
```

```python
sentences = re.split('。|；', "朱xx，女，2131-11-22就诊，检查项目为双侧乳腺X线检查。双侧乳腺实质呈致密型，腺体分布对称。右乳外上及外下象限局部腺体增厚。左侧乳腺内目前未见明确占位。双乳内可见点状及粗大钙化灶。皮肤未发现明显异常。双侧腋下可见淋巴结影。")
print(sentences)
tokenlist = jieba.lcut("双侧乳腺实质呈致密型，腺体分布对称")
print(tokenlist)


sentence_comma = re.split('，', "右侧乳腺中央区深部可见肿块，形态尚规则，边界模糊，大小约为30mm*29mm")
print(sentence_comma)

```

```python
# ++++++++++对每条病历进行处理

# 使用 with 语句管理文件，这样不需要手动 close，也不会报 "closed file" 的错
# 'w' 模式会覆盖写入，如果你想追加写入，请改为 'a'
with open(path, 'w', encoding='utf-8') as f:
    id = 0
    # 进行信息提取
    for patient in patients:  # 进入每条病历
        id += 1
        # 创建列表存放提取的结果（每次循环都要重置列表！）
        constitute = []  # 乳腺构成
        sides = []  # 肿块侧别
        quadrant = []  # 肿块象限
        depth = []  # 肿块深度
        size = []  # 肿块大小
        shape = []  # 肿块形状
        margin = []  # 肿块边界
        density = []  # 肿块密度
        calc_shape = []  # 肿块内钙化分布
        distort_sides = []  # 扭曲侧别

        # --- 你的提取逻辑保持不变 ---
        sentences = re.split('。|；', patient)
        for sentence in sentences[0:-1]:
            tokenlist = jieba.lcut(sentence)
            for word in tokenlist:
                if word in synonyms['乳腺构成']:
                    for token in tokenlist:
                        if token in exact['乳腺构成']:
                            constitute.append(token)
                if word in synonyms['结构扭曲']:
                    for token in tokenlist:
                        if token in exact['扭曲侧别']:
                            distort_sides.append(token)
                if word in synonyms['肿块判断']:
                    for token in tokenlist:
                        if token in exact['钙化形态']:
                            calc_shape.append((token))
                        if token in exact['肿块深度']:
                            depth.append(token)
                        if token in exact['肿块形状']:
                            shape.append(token)
                        if token in exact['肿块边缘']:
                            margin.append(token)
                        if token in exact['肿块密度']:
                            density.append(token)
                    # 提取肿块的侧别和象限信息
                    sentence_comma = re.split('，', sentence)
                    for item in sentence_comma:
                        tokenlist_comma = jieba.lcut(item)
                        for token_a in tokenlist_comma:
                            if token_a in synonyms['肿块判断']:
                                for token_b in tokenlist_comma:
                                    if token_b in exact['肿块侧别']:
                                        sides.append(token_b)
                                    if token_b in exact['肿块位置']:
                                        quadrant.append(token_b)
                            if token_a in exact['肿块大小']:
                                size1 = re.findall('(\d+mm[\*Xx]\d+mm|\d+mm)', item)
                                if size1: # 防止索引越界
                                    size1 = size1[0]
                                    size.append(size1)
                    break

        # --- 保存结果 ---
        template[id] = ({
            '乳腺构成': ' '.join(constitute),
            '肿块侧别': ' '.join(sides),
            '肿块象限和钟面位置': ' '.join(quadrant),
            '肿块深度': ' '.join(depth),
            '肿块大小': ' '.join(size),
            '肿块形状': ' '.join(shape),
            '肿块边缘': ' '.join(margin),
            '肿块密度': ' '.join(density),
            '肿块内钙化形态': ' '.join(calc_shape),
            '扭曲侧别': ' '.join(distort_sides)
        })

        # --- 写入文件 ---
        f.write(str(id) + '\n')
        for k, v in template[id].items():
            s2 = str(v)
            k2 = str(k)
            f.write(k2 + '：')
            f.write(s2 + '\n')

print("规则提取已完成，结果已保存至result_process.txt")
```

```python
import matplotlib.pyplot as plt
import os

def my_eval(label_value, predict_value, a):

    truepredict = 0
    foundguessed = 0
    foundcorrect = 0

    for i in range(a, len(label_value), 11):

        if label_value[i] == predict_value[i] and label_value[i] != '':
            truepredict += 1

        if label_value[i] != '':
            foundcorrect += 1

        if predict_value[i] != '':
            foundguessed += 1



    if foundguessed == 0:
        precision = 0.0
    else:
        precision = round(100 * truepredict / foundguessed, 2)
        
    if foundcorrect == 0:
        recall = 0.0
    else:
        recall = round(100 * truepredict / foundcorrect, 2)
        
    if precision + recall == 0:
        F1 = 0.0
    else:
        F1 = round(2 * precision * recall / (precision + recall), 2)
        
    return precision, recall, F1
```

```python
import matplotlib.pyplot as plt
import os
def eval(label_value, predict_value, a):
    # 计算需要的数据
    truepredict = 0  # 正确预测的值
    foundguessed = 0  # 被识别出的值
    foundcorrect = 0  # 标签中标定的值，非空
    for i in range(len(label_value)):
        if i == a:
            if label_value[i] == predict_value[i] and label_value[i] != '':
                truepredict += 1
            if label_value[i] != '':
                foundcorrect += 1
            if predict_value[i] != '':
                foundguessed += 1
            a = a + 11
    # 1. 计算 Precision
    if foundguessed == 0:
        precision = 0.0
    else:
        precision = round(100 * truepredict / foundguessed, 2)

    # 2. 计算 Recall
    if foundcorrect == 0:
        recall = 0.0
    else:
        recall = round(100 * truepredict / foundcorrect, 2)

    # 3. 计算 F1 (增加分母为0的判断)
    if precision + recall == 0:
        F1 = 0.0
    else:
        F1 = round(2 * precision * recall / (precision + recall), 2)
    return precision, recall, F1
```

```python
# -------------------计算总体数据precision、recall、F1------------
truepredict = 0 # 正确预测的值
foundguessed = 0 # 被识别出的值
foundcorrect = 0 # 标签中标定的值，非空
for i in range(len(label_value)):
    if label_value[i] == predict_value[i] and label_value[i] != '':
        truepredict += 1
    if label_value[i] != '':
        foundcorrect += 1
    if predict_value[i] != '':
        foundguessed += 1
precision = round(100 * truepredict / foundguessed, 2)
recall = round(100 * truepredict / foundcorrect, 2)
F1 = round(2 * precision * recall / (precision + recall), 2)

# -------------------逐条属性评估precision、recall、F1------------
precision_a, recall_a, F1_a = eval(label_value, predict_value, 1)
precision_b, recall_b, F1_b = eval(label_value, predict_value, 2)
precision_c, recall_c, F1_c = eval(label_value, predict_value, 3)
precision_d, recall_d, F1_d = eval(label_value, predict_value, 4)
precision_e, recall_e, F1_e = eval(label_value, predict_value, 5)
precision_f, recall_f, F1_f = eval(label_value, predict_value, 6)
precision_g, recall_g, F1_g = eval(label_value, predict_value, 7)
precision_h, recall_h, F1_h = eval(label_value, predict_value, 8)
precision_i, recall_i, F1_i = eval(label_value, predict_value, 9)
precision_j, recall_j, F1_j = eval(label_value, predict_value, 10)

# -------------------打印评估结果------------
print("all precision : %s; recall : %s; F1 : %s \n" % (precision, recall, F1))
print("constitute precision : %s; recall : %s; F1 : %s \n" % (precision_a, recall_a, F1_a))
print("lump_sides precision : %s; recall : %s; F1 : %s \n" % (precision_b, recall_b, F1_b))
print("lump_quadrant precision : %s; recall : %s; F1 : %s \n" % (precision_c, recall_c, F1_c))
print("depth precision : %s; recall : %s; F1 : %s \n" % (precision_d, recall_d, F1_d))
print("size precision : %s; recall : %s; F1 : %s \n" % (precision_e, recall_e, F1_e))
print("shape precision : %s; recall : %s; F1 : %s \n" % (precision_f, recall_f, F1_f))
print("margin precision : %s; recall : %s; F1 : %s \n" % (precision_g, recall_g, F1_g))
print("density precision : %s; recall : %s; F1 : %s \n" % (precision_h, recall_h, F1_h))
print("calc_shape precision : %s; recall : %s; F1 : %s \n" % (precision_i, recall_i, F1_i))
print("distort_sides precision : %s; recall : %s; F1 : %s \n" % (precision_j, recall_j, F1_j))

```
